# Class Distribution

Stratified class distribution across train/val/test splits.

```python
import json
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent / "src"))
import torch
from utils import SPLITS_DIR, DATA_DIR, AAMI_CLASSES

data = torch.load(DATA_DIR / "processed_mlii.pt", map_location="cpu", weights_only=True)
y = data["y"].numpy()
for name in ["intra_patient", "inter_patient"]:
    with open(SPLITS_DIR / f"{name}.json") as f:
        s = json.load(f)
    for part in ["train", "val", "test"]:
        counts = [(y[s[part]] == c).sum() for c in range(len(AAMI_CLASSES))]
        print(f"{name} {part}: {dict(zip(AAMI_CLASSES, counts))}")
```
