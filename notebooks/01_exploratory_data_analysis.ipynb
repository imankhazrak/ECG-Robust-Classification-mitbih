# Exploratory Data Analysis

Load processed MIT-BIH beats and inspect signal statistics and basic visualizations.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent / "src"))
import torch
from utils import DATA_DIR, AAMI_CLASSES

data = torch.load(DATA_DIR / "processed_mlii.pt", map_location="cpu", weights_only=True)
X, y = data["X"], data["y"]
print("Shape:", X.shape, "Labels:", y.shape)
print("Class counts:", [(c, (y == i).sum().item()) for i, c in enumerate(AAMI_CLASSES)])
```
